```python
from utils import config, log, miscellaneous, seed
import os
import numpy as np
import basic_trainer
from NeuroMax.NeuroMax import NeuroMax
import evaluations
import datasethandler
import scipy
import torch
import wandb
from time import time
from sentence_transformers import SentenceTransformer

RESULT_DIR = 'results'
DATA_DIR = 'datasets'
    # if args.dataset in ['YahooAnswers']:
    #     read_labels = True
    # else:
    #     read_labels = False
read_labels = True
plm_model = SentenceTransformer("all-MiniLM-L6-v2", device="cuda")

    # load a preprocessed dataset
dataset = datasethandler.BasicDatasetHandler(
        os.path.join(DATA_DIR, "AGNews"), device="cuda", read_labels=read_labels,
        as_tensor=True, contextual_embed=True)

    # create a model
pretrainWE = scipy.sparse.load_npz(os.path.join(
        DATA_DIR, "AGNews", "word_embeddings.npz")).toarray()

    # model = NeuroMax(vocab_size=dataset.vocab_size,
    #                 num_topics=args.num_topics,
    #                 num_groups=args.num_groups,
    #                 dropout=args.dropout,
    #                 pretrained_WE=pretrainWE if args.use_pretrainWE else None,
    #                 weight_loss_GR=args.weight_GR,
    #                 weight_loss_ECR=args.weight_ECR,
    #                 alpha_ECR=args.alpha_ECR,
    #                 alpha_GR=args.alpha_GR,
    #                 weight_loss_InfoNCE=args.weight_InfoNCE,
    #                 beta_temp=args.beta_temp)
model = NeuroMax(vocab_size=dataset.vocab_size, 
                     num_topics=50,
                    num_groups=20, 
                    hidden_dim_1=384, 
                    hidden_dim_2=384, 
                    embed_size=dataset.contextual_embed_size, 
                    dropout=0.1, 
                    theta_temp=1.0,
                    DT_alpha=3.0, TW_alpha=2.0,
                    weight_loss_GR=1,
                    alpha_GR=1, 
                    weight_loss_InfoNCE=150)
model.weight_loss_GR = 1
model = model.to("cuda")

def get_doc_simple_embedidng(input_bow):
    data_size = input_bow.shape[0]
    doc_simple_embedding = list()
    all_idx = torch.split(torch.arange(data_size), 200)

    with torch.no_grad():
        model.eval()
        for idx in all_idx:
            batch_input = input_bow[idx]
            batch_embedding = model.simpleembedding(batch_input)
            doc_simple_embedding.extend(batch_embedding.cpu().tolist())

    doc_simple_embedding = torch.tensor(doc_simple_embedding).to("cuda")
    return doc_simple_embedding

start_train_inference = time()        
train_simple_embedding = get_doc_simple_embedidng(dataset.train_data)
        #self.train_simple_embedding = dataset_handler.train_contextual_embed
train_theta = model.get_theta(train_simple_embedding, train_simple_embedding)
end_train_inference = time()
print("*******************")
print("Train inference time: " + str(end_train_inference - start_train_inference))
print("*******************")
```
